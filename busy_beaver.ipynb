```python
def replace(string, i, c):
    return string[:i] + c + string[i + 1:]

def halts(tm):
    # Simulates Turing Machine tm and checks if it halts in <= 10 steps.
    state = ("q0", "0" * 21, 10)
    for i in range(10):
        q = state[0]
        tape = state[1]
        position = state[2]
        inp = tape[position]
        transition = tm[(q, inp)]
        # Update q
        q = transition[0]
        # Update tape
        tape = replace(tape, position, transition[1])
        # Update position
        if (transition[2] == "L"):
            position -= 1
        else:
            position += 1
        # Update state
        state = (q, tape, position)
        if (q == "accept" or q == "reject"):
            return (True, i + 1)
    return (False, 0)

states = ["q0", "q1", "accept"]
symbols = ["0", "1"]
moves = ["L", "R"]
largest_number_of_steps = 0
machine = None
for s0 in states:
    for s1 in states:
        for s2 in states:
            for s3 in states:
                for symb0 in symbols:
                    for symb1 in symbols:
                        for symb2 in symbols:
                            for symb3 in symbols:
                                for m0 in moves:
                                    for m1 in moves:
                                        for m2 in moves:
                                            for m3 in moves:
                                                tm = {
                                                        ("q0", "0") : (s0, symb0, m0),
                                                        ("q0", "1") : (s1, symb1, m1),
                                                        ("q1", "0") : (s2, symb2, m2),
                                                        ("q1", "1") : (s3, symb3, m3)
                                                }
                                                does_halt, steps = halts(tm)
                                                if (does_halt and steps > largest_number_of_steps):
                                                    largest_number_of_steps = steps
                                                    machine = tm

print(largest_number_of_steps)
print(machine)
```
